## Finding the Benching percentage of this company

```python
count = df.value_counts('EverBenched')
num_of_benched = count.Yes
num_of_emp = df.shape[0] #num of rows represents num of employees

benched_percent = (num_of_benched / num_of_emp) * 100

benched_percent
```

```python
df_benched = df.loc[df['EverBenched'] == "Yes"]
df_benched
total_benched = df_benched.shape[0]
total_employee = df.shape[0]

percentage = (total_benched / total_employee) * 100
print(f"The benching percentage of this company is {percentage:.2f}%")
```
